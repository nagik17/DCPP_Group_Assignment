```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pyplot
import seaborn as sns
```

```python
data1 = pd.read_csv('/content/Receipe_Details.csv')
data2 = pd.read_csv('/content/additional_receipes.csv')
data = data1.append(data2)
data4 = pd.read_excel('/content/SanjeevKapoor.xlsx',sheet_name='in')
```

```python
column_names = ['Recipe_Name', 'Total_Time', 'Cuisine','Ingredients_List',  'Number_of_Ingredients', 'Prep_Time', 'Cook_Time', 'Serves', 'Level_of_Cooking' , 'Course']
final_data=pd.DataFrame(columns = column_names)
```

```python
#removing recipe card at the end of each recipe name
final_data['Recipe_Name'] = data['Unnamed: 0']
final_data['Recipe_Link'] = data4
```

```python
recipe_name=[]
for i in final_data['Recipe_Name']:
  recipe_name.append(' '.join((i.split(' ')[0:-2])))
final_data['Recipe_Name']=recipe_name
data3 = pd.read_excel('/content/Recipedetail_updated.xlsx', sheet_name ='in' )
#final_data['Taste'] = data3['Taste']
url_data = pd.read_excel('/content/SanjeevKapoorimages - Copy.xlsx',sheet_name = 'in')
final_data['Image_URL'] = url_data
```

```python
list_of_main_ing = []
import ast
import re
for i in data['0']:
  i = ast.literal_eval(i)
  temp2=[]
  temp = i['Main Ingredients']
  list_of_main_ing.append(temp)
print(list_of_main_ing)
```

```python
cuisine_list=[]; course_list =[]
for i in data['0']:
  i = ast.literal_eval(i)
  temp = i['Cuisine']
  cuisine_list.append(temp)
  temp = i['Course']
  course_list.append(temp)
final_data['Cuisine'] = cuisine_list
final_data['Course'] = course_list
```

```python
prep_time=[];cook_time=[];loc_list=[];serve_list=[]; tot_time=[]
for i in data['1']:
  i = ast.literal_eval(i)
  prep = i['Prep Time']
  cook = i['Cook time']
  if '-' in prep:
    if 'minutes' in prep:
      temp = prep.split('-')
      prep_time.append(float(temp[0]))
    elif 'day' in prep:
      temp = prep.split('-')
      prep_time.append(24*60*float(temp[0]))
    elif 'hour' in prep:
      temp = prep.split('-')
      prep_time.append(60*float(temp[0]))
    else:
      prep_time.append(0)
  else:
    prep_time.append(0)
      
  if '-' in cook:
    if 'minutes' in cook:
      temp = cook.split('-')
      cook_time.append(float(temp[0]))
    elif 'day' in cook:
      temp = cook.split('-')
      cook_time.append(24*60*float(temp[0]))
    elif 'hour' in cook:
      temp = cook.split('-')
      cook_time.append(60*float(temp[0]))
    else:
      cook_time.append(0)
  else:
      cook_time.append(0)
  try:
    tot_time.append(prep_time[-1]+cook_time[-1])
  except:
    tot_time.append(0)
    
  temp = i['Serve']
  serve_list.append(temp)
  temp = i['Level Of Cooking']
  loc_list.append(temp)
final_data['Prep_Time'] = prep_time
final_data['Total_Time'] = tot_time
final_data['Cook_Time'] = cook_time
final_data['Serves'] = serve_list
final_data['Level_of_Cooking'] = loc_list
final_data.head(3)
```

```python
recipe_links = []
for j,i in final_data.iterrows():
  if 'sanjeevkapoor.com' in i['Recipe_Link']:
    str1 = 'https://www.sanjeevkapoor.com/Recipe/'
    temp = i['Recipe_Name'].split(' ')
    name = '-'.join(temp)
    str2 = '.html'
    recipe_links.append(str1+str(name)+str2)
  else:
    recipe_links.append(i['Recipe_Link'])
recipe_links
```

```python
img_links = []
for j,i in final_data.iterrows():
  if 'sanjeevkapoor.com' in i['Recipe_Link']:
    if 'SK Khazana' in i['Recipe_Name']:
      str1 = 'https://www.sanjeevkapoor.com/UploadFiles/RecipeImages/'
      temp = i['Recipe_Name'].split(' ')
      name = '-'.join(temp)
      str2 = '.jpg'
    else:
      str1 = 'https://www.sanjeevkapoor.com/UploadFiles/RecipeImages/'
      temp = i['Recipe_Name'].split(' ')
      name = '-'.join(temp)
      str2 = 'skk.jpg'
    img_links.append(str1+str(name)+str2)
  else:
    img_links.append(i['Image_URL'])
img_links
```

```python
final_data['Recipe_link'] = recipe_links
final_data['Image_URL'] = img_links
final_data['Ingredients_List'] = data['2']
no_of_ing = [len(i.split(',')) for i in data['2']]
#print(no_of_ing)
```
